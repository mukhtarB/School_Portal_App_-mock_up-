### admin

```python
def admin(Name):
    print(f'Welcome {Name}')
    q = input('Would you like to:\n a - search an entry\n b - delete an entry\n c - verify student registrations\n\
 d - print list of registered students and details').strip()
    
    if q.casefold() == 'a':
        a_ID = input('Enter student ID:').strip()
        a_ID = a_ID.replace('/', '_')
        if os.path.isfile(f'school_app_db\\{a_ID}.txt'):
            with open(f'school_app_db\\{a_ID}.txt', mode = 'r') as search_res:
                for line in search_res:
                    print(line)
        else:
            print('Student not registered')
    elif q.casefold() == 'b':
        a_ID = input('Enter student ID:').strip()
        a_ID = a_ID.replace('/', '_')
        if os.path.isfile(f'school_app_db\\{a_ID}.txt'):
            delete = input(f'ARE YOU SURE YOU WANT TO DELETE USER {a_ID} :').strip()
            if delete.casefold() == 'yes':
                os.remove(f'school_app_db\\{a_ID}.txt')
                print('SUCCESSFULLY DELETED!')
        else:
            print('Student Not Registered')
    elif q.casefold() == 'c':
        a_ID = input('Enter student ID:').strip()
        a_ID = a_ID.replace('/', '_')
        if os.path.isfile(f'school_app_db\\{a_ID}.txt'):
            with open(f'school_app_db\\{a_ID}.txt', mode = 'r') as verify_reg:
                verify_reg = verify_reg.readlines()
            for line in verify_reg:
                    print(line)
                
            if len(verify_reg) >= 5:
                q2 = input('accept? or deny?')
                if q2.casefold() == 'accept' or q2.casefold() == 'accepted':
                    verify_reg[-1] = f'                            Registration Status : Accepted\n'
                elif q2.casefold() == 'deny' or q2.casefold() == 'denied':
                    verify_reg[-1] = f'                            Registration Status : Denied\n'
                else:
                    pass
                
                with open(f'school_app_db\\{a_ID}.txt', mode = 'w') as u_reg:
                    u_reg.write(''.join(verify_reg))
            else:
                print('Student is yet to register courses')
        else:
            print('ID not registered')
    elif q.casefold() == 'd':
        with open(f'school_app_db\\database_all.txt', mode = 'r') as student_list:
            for line in student_list:
                print(line)
            
            
# admin('Prof. John Travolta')
```

```python
def reg_courses(a_ID, Name):
    '''
    DOCSTRING: Function that reg student courses
    INPUT:
    OUTPUT:
    '''
    if os.path.isfile(f'school_app_db\\{a_ID}.txt'):
        with open(f'school_app_db\\{a_ID}.txt', mode = 'r') as checkreg:
            det = checkreg.readlines()
            
        if not 'Course REG:\n' in det:
            course_Avail = ['Maths', 'Bio', 'Chem', 'Agric', 'Comp sci', 'Physics', 'Stats']

            print("\nAVAILABLE COURSES:", course_Avail)
            courses = input('-Select a course from above (min = 2, max = 5)\n-Split each with a " , " only\n\n')

            course_L = courses.split(',')
            course_L2 = []
            for i in course_L:
                i = i.strip()
                i = i.capitalize()
                course_L2.append(i)
            course_list2 = set(course_L2)
            course_list = list(course_list2)
            
            for i in course_list:
                c_details = course_list

                if i in course_Avail:
                    if len(course_list) >= 2 and len(course_list) <= 5:
                        print(f'Congratulations, {Name} you succesfully registered {i}\n')
                        c_details = ', '.join(c_details)

                    else:
                        print(f'Sorry {Name}, REGISTRATION FAILED!\nEnsure you have min = 2 and max = 5 courses.')
                        c_details = ""
                        break

                else: 
                    print(f'\nUnable to register "{i}"\nNot part of available courses.\n')
                    print(f' » REGISTRATION FAILED!\n')
                    c_details = ""
                    break
            
            # Updating database
            #---------------------------------
            if c_details:
                cur_timestamp = datetime.datetime.now()
                timestamp = cur_timestamp.strftime("%c")
                x = 0           
                while x < 1:
                    if os.path.isfile(f'school_app_db\\{a_ID}.txt'):
                        with open(f'school_app_db\\{a_ID}.txt', 'a') as courseList:
                            courseList.write(f'\n-----------\nCourse REG:\nLog: Registered - {timestamp}.\n\
                            Courses : {c_details}.\n\
                            Registration Status : Pending\n') #update user info

                    if os.path.isfile(f'school_db_backup\\{a_ID}.txt'):
                        with open(f'school_db_backup\\{a_ID}.txt', 'a') as courseList:
                            courseList.write(f'\n-----------\nCourse REG:\nLog: Registered - {timestamp}.\n\
                            Courses : {c_details}.\n\
                            Registration Status : Pending\n') #back up
                    x += 1
                    
             #done!
             #--------------------------------
            
                a3 = input('Would you like to take classes?:').casefold()
                if a3 == 'yes':
                    take_class(a_ID, Name)
            
            return c_details
        else:
            print(f'Sorry {Name}, you already registered your courses.')
    else:
        print(f'Cannot Register, user {a_ID} does not exist')
# reg_courses('Tobi', '9999_tt')
```

```python
#02. Authentication to view data - Calling  data from the database
def authentication():
    '''
    DOCSTRING: Function that authenticates users (students and admin)
    INPUT:
    OUTPUT:
    '''
    print('\n---------WELCOME, please input your login details')
    a_ID = input('\nEnter your ID:\n').casefold()
    a_ID = a_ID.replace('/', '_')

    password = input('\nEnter your password: -Case Sensitive-\n')
    
    db = os.path.isfile(f'school_app_db\\{a_ID}.txt')
    if db:
        with open(f'school_app_db\\{a_ID}.txt', mode = "r") as x:
            details = x.readlines()

        Name = details[3].strip()[7:-1]
        det_pd = details[4].strip()[11:-1]
        secure_pwd =  det_pd[6::7]

        if password == secure_pwd:
            if a_ID.endswith('@school.edu.ng'):
                admin(Name)
            else:
                print(f'\n\r-----------\n\rWelcome, {Name}')

                print(f'\nHere are your details:')
                for itx in details:
                    print(itx)
                
                reg_c = input('Would you like to register courses or take classes?\n\
                respond with either "courses" or "classes"').casefold()
                if reg_c == 'courses':
                    reg_courses(a_ID, Name)   #call funct reg courses
                else:
                    if len(details) >= 6:
                        take_class(a_ID, Name) #call funct take classes
                    else:
                        print('\nYou have no registered course.')
                        wr = input('\nREGISTER NOW? ').casefold()
                        if wr == 'yes':
                            reg_courses(a_ID, Name)
                        else:
                            print('\nThe system would be logging you out now.\nDo take care.')
                            log_out()


        else:
            print('Wrong ID or Password.')
    else:
        print('Wrong ID or password.')
        regYY = input('Dont have an acc with us? Would you like to register?').casefold()
        if regYY == 'yes':
            reg_01()
        else:
            log_out()
# authentication()
```

### STEP 2 of registration

```python
def reg_02(Uname):
    '''
    DOCSTRING: SECT 2 - Function registrating user to database.
    INPUT:Uname
    OUTPUT:password, id and _id
    '''
    pd = input(f'\nPlease {Uname} enter your password: -Case Sensitive-\n')
    
#     -----------------------
    def hash_pd(pd):
        '''
        DOCSTRING: Function hashing the pd into a secure pd.
        INPUT:pd
        OUTPUT:secured_pd
        '''
        encode = ['a','s','d','f','j','k','l','q','/','?','%','_','=','0','1','2','4','6','8','A','S','D','F','J','K','L','Q']

        encoded = ""
        if pd:
            for it in pd:
                g6 = ''.join(random.sample(encode, 6))
                it = g6 + it
                encoded += it
                
        print('---------')
        return encoded

    pword = hash_pd(pd)
#     --------------------

    ID = input(f'\nPlease {Uname} give your ID:\n').casefold()
    if ID.endswith('@school.edu.ng'):
        title = input('Rank / Title : ').strip()
        Uname = title + ' ' + Uname
    _ID = ID.replace('/', '_')
    
    if ID and Uname and pword:
        cur_timestamp = datetime.datetime.now()
        timestamp = cur_timestamp.strftime("%c")
        
        info = f"» » {ID}:\n\
                    Matric No : {ID},\n\
                    Name : {Uname},\n\
                    Password : {pword}"
        print('=======» Portal Registration Successful')
        print(info)
        

        ii=0
        while ii < 1:
            with open(f'school_app_db\\database_reg.txt', mode = "a") as x4:
                x4.write(f'\n{Uname} - successful\n')
            with open(f'school_db_backup\\database_reg.txt', mode = "a") as x5:
                x5.write(f'\n{Uname} - successful\n')
            
            #Updating database w results
            #-------------------
            if not ID.endswith('@school.edu.ng'):
                if os.path.isfile(f'school_app_db\\database_all.txt'):
                    with open('school_app_db\\database_all.txt', 'a') as db_txt:
                        db_txt.write(f'\n\r {info}.\nLog: Registered to portal - {timestamp}.\n')

                if os.path.isfile(f'school_db_backup\\database_all.txt'):
                    with open('school_db_backup\\database_all.txt', 'a') as db_txt:
                        db_txt.write(f'\n\r {info}.\nLog: Registered to portal - {timestamp}.\n')
                    
            #--------------------
            if not os.path.isfile(f'school_app_db\\{_ID}.txt'):
                with open(f'school_app_db\\{_ID}.txt', 'x') as uxer_txt:
                    uxer_txt.write(f'Log: Created - {timestamp}.\n{info}.\n')
                    
            if not os.path.isfile(f'school_db_backup\\{_ID}.txt'):
                with open(f'school_db_backup\\{_ID}.txt', 'x') as uxer_txt:
                    uxer_txt.write(f'Log: Created - {timestamp}.\n{info}.\n')
            ii+=1
            
            #done!
            #--------------------
            
        authYY = input('Would you like to login?').casefold()
        if authYY == 'yes':
            authentication()
            
    else:
        print('Registration failed. Ensure you fill in all required fields.')
        info = ""
```

### STEP 1 of registration

```python
def reg_01():
    '''
    DOCSTRING: SECT 1 - Function handling session mgmt on portal registration.
    INPUT:
    OUTPUT:
    ''' 
    
    print('--------\nFill the form to register with us.\n')
    Uname = input('Please give your name:\n').capitalize()

    if os.path.isfile(f'school_app_db\\database_reg.txt'):
        with open(f'school_app_db\\database_reg.txt', mode = "r") as xx:
            attempted = xx.readlines()
            
        if f'{Uname} - successful\n' in attempted:
            print('\nYou have successfully registered,\n Login to view your details.\n')
            step2 = authentication()
        
        elif f'{Uname}\n' in attempted or f'{Uname}' in attempted:
            print('\nYou seem to have had a problem with your previous registration. Please try again.')
            step2 = reg_02(Uname)
        
        else:
            print(f'\nWelcome {Uname}, you seem to be a first timer, proceed to stage 2 of registeration.')
            with open(f'school_app_db\\database_reg.txt', mode = "a") as xxx:
                xxx.write(f'\n{Uname}')
            step2 = reg_02(Uname)

            
            
    return [Uname, step2]
    
```

```python
def log_out():
    '''
    DOCSTRING: On log out, this runs
    INPUT: None
    OUTPUT: None
    '''
    ans = input('==========\nHello, Would you like to register or login\n R for register, L for Login: ').upper()
    if ans == 'R':
        reg_01()
    elif ans == 'L':
        authentication()
    else:
        print('Invalid params')
log_out()
```
